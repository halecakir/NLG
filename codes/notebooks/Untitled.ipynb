```python
MATCHED_LIBS_DIR = "/data/huseyinalecakir_data/predictedLibsREAD_CONTACTS"

import os
import json
from collections import defaultdict
libs = defaultdict(int)
libs_with_dir = defaultdict(list)

for file in os.listdir(MATCHED_LIBS_DIR):
    path = os.path.join(MATCHED_LIBS_DIR, file)
    if path.endswith(".json"):
        with open(path) as target:
            data = json.load(target)
            for lib in data["libraries"]:
                libs[lib["name"]] += 1
                for loc in lib["matched_root_package"]:
                    libs_with_dir[lib["name"]].append((loc, lib["similarity"]))

                    sorted_libs = sorted(libs.items(), key=lambda t:t[1], reverse=True)
for lib, count in sorted_libs:
    print(lib, count)
"""for lib, locations in libs_with_dir:
    print(lib)
    for loc in locations:
        print(loc, end=" ")"""
```
